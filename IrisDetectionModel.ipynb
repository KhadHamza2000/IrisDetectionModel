```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0, test_size=0.25)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
import numpy as np
X_new = np.array([[1,3.9,4,0.5]])
prediction = knn.predict(X_new)
print(iris_dataset['target_names'][prediction])
```

```python
y_pred = knn.predict(X_test)
print(f"Test set predictions:\n{y_pred}")
```

```python
print(f"Test set score: {np.mean(y_pred == y_test):.2f}")
```

```python
print(f"Test set score: {knn.score(X_test, y_test)}")
```
